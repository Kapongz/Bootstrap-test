# 10. ເພີ່ມ Grade ຄິດໄລ່ຈາກ Total ເງຶື່ອນໄຂຄະແນນຫຼາຍກ່ວາ 90 ສະແດງ Exellent ຄະແນນຫຼາຍກ່ວາ 70 ສະແດງ Better ຄະແນນຫຼາຍກ່ວາ 50 ສະແດງ Good ຄະແນນຕ່ຳກ່ວາ 50 ສະແດງ Fail   

---



```python

conditions = [


    (set_zero['Total'] < 50),

    (set_zero['Total'] >= 50) & (set_zero['Total'] < 70),

    (set_zero['Total'] >= 70) & (set_zero['Total'] < 90),

    (set_zero['Total'] >= 90)


]
letters = ['Fail', 'Good', 'Better', 'Excellent']
set_zero['Grade'] = np.select(conditions, letters)
set_zero
```

# 10. ສະແດງຊື່ທີ່ໄດ້ Exellent

---



```python
set_zero.loc[set_zero['Grade'] == "Excellent", 'Name']
```
